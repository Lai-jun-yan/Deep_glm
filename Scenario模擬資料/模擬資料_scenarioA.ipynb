```python
n = 500 # sample size
p = 50 # predictor數量(J)
rho = 0.5 # predictor相關程度
scenario = "A"
```

```python
import numpy as np

def build_covariance(p, rho):
    Sigma = np.zeros((p, p))

    for i in range(p):
        for j in range(p):
            Sigma[i, j] = rho ** abs(i-j)

    return Sigma
```

```python
Sigma = build_covariance(5,0.5)

print(Sigma)
```

```python
def generate_X(n, p, rho, seed=None):

    if seed is not None:
        np.random.seed(seed)

    Sigma = build_covariance(p, rho)

    X = np.random.multivariate_normal(
        mean=np.zeros(p),
        cov=Sigma,
        size=n
    )

    return X
```

```python
X = generate_X(
    n=500,
    p=50,
    rho=0.5,
    seed=123
)
```

```python
def standardize(X):

    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0, ddof=0)

    return (X - mean) / std
```

```python
X = standardize(X)
```

```python
beta = np.zeros(50)

beta[:5] = [2,1.5,-2,1,-1]
```

```python
mu = X @ beta
```

```python
sigma = 1

epsilon = np.random.normal(
    0,
    sigma,
    size=n
)
```

```python
Y = mu + epsilon
```
