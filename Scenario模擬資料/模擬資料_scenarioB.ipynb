```python
n = 500 # sample size
p = 50 # predictor數量(J)
rho = 0.5 # predictor相關程度
scenario = "B"
```

```python
import numpy as np

def build_covariance(p, rho):
    idx = np.arange(p)
    return rho ** np.abs(np.subtract.outer(idx, idx))

def generate_X(n, p, rho):

    Sigma = build_covariance(p, rho)

    X = np.random.multivariate_normal(
        mean=np.zeros(p),
        cov=Sigma,
        size=n
    )

    return X
```

```python
X = generate_X(n,p,0.5)
```

```python
hidden = 20

weight_scale = 0.3

W1 = np.random.normal(0,weight_scale,(p,hidden))

W2 = np.random.normal(0,weight_scale,(hidden,p))
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
H = relu(
    X @ W1
)
```

```python
logits = H @ W2
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
alpha = sigmoid(logits)
```

```python
Z = alpha * X
```

```python
beta = np.zeros(p)

beta[:5] = [2,-2,1.5,1,-1]
```

```python
mu = Z @ beta
```

```python
sigma = 1

eps = np.random.normal(
    0,
    sigma,
    n
)
```

```python
Y = mu + eps
```
