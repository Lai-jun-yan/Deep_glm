```python
X = data.iloc[:,0:50].values
Y = data.iloc[:,50].values

X = X[:,:,None] # 變tensor

print(X.shape)
print(Y.shape)
```

```python
d_k = 8 # 1維的embedding要先被映射到8維

Wq = np.random.randn(1,d_k)

Wk = np.random.randn(1,d_k)

Wv = np.random.randn(1,d_k)
```

```python
Q = X @ Wq
K = X @ Wk
V = X @ Wv
```

```python
scores = np.matmul(
    Q,
    K.transpose(0,2,1)
)
```

```python
def softmax(x):
    exp_x=np.exp(
        x-np.max(x,axis=-1,keepdims=True)
    )
    return exp_x / np.sum(
        exp_x,
        axis=-1,
        keepdims=True
    )
```

```python
attention_weights = softmax(
    scores/np.sqrt(d_k)
)
```

```python
attention_output = np.matmul(
    attention_weights,
    V
)
```

```python
attention_output = attention_output.reshape(
    500,
    -1
)
```
