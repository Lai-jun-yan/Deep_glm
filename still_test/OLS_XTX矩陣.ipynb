```python
import pandas as pd
import numpy as np

# 先讀進模擬資料
data = pd.read_csv(r"C:\Users\USER\Desktop\碩論\程式碼\embedding_data.csv")

cols = ["X2","X3","X1"] # 針對變數標準化，後面做softmax的時候，數值才不會爆掉

data[cols] = (data[cols] - data[cols].mean()) / data[cols].std()

whole = data.copy()

data = data.iloc[0:8,:]

validation = whole.iloc[70:100,:]

print(len(whole))
print(len(data))
print(len(validation))
print(data)
```

```python
X = data[["X2","X3","X1"]]

print(X)
```

```python
XTX = X.T @ X

print(XTX)
```

```python
y = data["Y"].head(8)

print(y)
```

```python
# beta
beta = np.linalg.inv(X.T @ X) @ X.T @ y

print(beta)
```

```python
# 用套件驗證
import statsmodels.api as sm

X = data[["X2","X3","X1"]].head(8)

model = sm.OLS(
    data["Y"].head(8),
    X          # 不加 constant
)

result = model.fit()

print(result.params)
```
